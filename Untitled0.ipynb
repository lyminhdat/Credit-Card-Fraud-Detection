**Tách các giá trị X và Y**

```python
X = data.drop(['Class'], axis = 1)
Y = data['Class']
print(X.shape)
print(Y.shape)
# Lấy các giá trị chỉ để xử lý
# (đây là một mảng numpy không có các cột)
xData = X.values
yData = Y.values
```

**Đào tạo và kiểm tra phân nhánh dữ liệu**

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(xData, yData, test_size = 0.2, random_state=42)
```

**Xây dựng mô hình rừng ngẫu nhiên**

```python
# Xây dựng Bộ phân loại Rừng Ngẫu nhiên (RANDOM FOREST)
from sklearn.ensemble import RandomForestClassifier
# tạo mô hình rừng ngẫu nhiên
rfc = RandomForestClassifier()
rfc.fit(xTrain, yTrain)
# predictions
yPred = rfc.predict(xTest)
```

```python
# Đánh giá bộ phân loại
# in ra mọi điểm số của bộ phân loại
# chấm điểm trên bất kỳ khía cạnh nào
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score, matthews_corrcoef
from sklearn.metrics import confusion_matrix
```

```python
n_outliers =len(fraud)
n_errors = (yPred != yTest).sum()
print("The model used is Random Forest Classifier")

# Độ chính xác: Tỷ lệ dự đoán đúng trên tổng số dự đoán.
acc = accuracy_score(yTest, yPred)
print("The accuracy: {}".format(acc))

# Độ chính xác: Tỷ lệ các dự đoán đúng trong số các dự đoán được dự đoán là đúng
prec = precision_score(yTest, yPred)
print("The precision is: {}".format(prec))

# Độ thu hồi
rec = recall_score(yTest, yPred)
print("The recall is {}".format(rec))

# Điểm f1
f1 = f1_score(yTest, yPred)
print("The F1 score is {}".format(f1))

# Hệ số tương quan Matthews là
MCC = matthews_corrcoef(yTest, yPred)
print("The Matthews correllation coefficient is {}".format(MCC))
```

```python
LABELS = ['Normal', 'Fraud']
conf_matrix = confusion_matrix(yTest, yPred)
plt.figure(figsize=(8, 8))
sns.heatmap(conf_matrix, xticklabels=LABELS, yticklabels=LABELS, annot=True, fmt="d")
plt.title("Confusion matrix")
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```
